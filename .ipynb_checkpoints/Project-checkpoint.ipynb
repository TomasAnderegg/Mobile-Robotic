```python
from tdmclient import ClientAsync
# import tdmclient.notebook
# await tdmclient.notebook.start()
import time

import matplotlib.pyplot as plt
import numpy as np
client = ClientAsync()
node = await client.wait_for_node()

await node.lock()
```

Constantes

```python
THYMIO_MMS = 0.2978
D_BASELINE = 95 #mm
```

Odom√©trie

```python
async def odometrie(x,y,theta,motor_l,motor_r,dt):
   
    v = (motor_r + motor_l)*THYMIO_MMS/2
    w = (motor_r - motor_l)*THYMIO_MMS

    theta += w/D_BASELINE
    
    x += v*dt*np.cos(theta)
    y += v*dt*np.sin(theta)

    return x,y, theta
```

```python
def motors(left, right):
    return {
        "motor.left.target": [left],
        "motor.right.target": [right],
    }
```

Main

Kalmann

```python
def Jacobien(theta, v_right, v_left):
    global thymio_speed_to_mms, wheel_base, dt
    """
    Computes the Jacobian matrix of the odometry model, 
    """
    alpha = thymio_speed_to_mms * (v_right + v_left) / 2
    beta = thymio_speed_to_mms * (v_right - v_left) / wheel_base

    A = - (alpha * dt * np.sin(theta + dt * beta))
    B = (1/2) * dt * np.cos(theta + dt * beta) - (alpha * (1 / wheel_base) * dt**2 * np.sin(theta + dt * beta))
    C = (1/2) * dt * np.cos(theta + dt * beta) + (alpha * (1 / wheel_base) * dt**2 * np.cos(theta + dt * beta))
    D = alpha * dt * np.cos(theta + dt * beta)
    E = (1/2) * dt * np.sin(theta + dt * beta) + (alpha * (1 / wheel_base) * dt**2 * np.cos(theta + dt * beta))
    F = (1/2) * dt * np.sin(theta + dt * beta) - (alpha * (1 / wheel_base) * dt**2 * np.cos(theta + dt * beta))
    G = dt/wheel_base
    H = -dt/wheel_base
    
    J = np.array([[1, 0, A, B, C],
                  [0, 1, D, E, F],
                  [0, 0, 1, G, H],
                  [0, 0, 0, 1, 0],
                  [0, 0, 0, 0, 1]])
    
    return J
```
