```python
from tdmclient import ClientAsync
import numpy as np

client = ClientAsync()
node = await client.wait_for_node()
await node.lock()
```

```python
THYMIO_MMS = 0.3183
D_BASELINE = 95 #mm
```

```python
def motors(left, right):
    return {
        "motor.left.target": [left],
        "motor.right.target": [right],
    }
```

```python
async def odometrie(x,y,theta,motor_l,motor_r,dt):
   
    v = (motor_r + motor_l)*THYMIO_MMS/2
    w = (motor_r - motor_l)*THYMIO_MMS

    theta += w*dt/D_BASELINE
    
    x += v*dt*np.cos(theta)
    y += v*dt*np.sin(theta)

    return x,y, theta
```

```python
x = 0 
y= 0
theta = 0
state = False

x_odom = []
y_odom = []
theta_odom = []
v_r = []
v_l = []

count = 100 #donc apr√®s 10s le robot s'arrete

async def main():
    
    global x,y,theta, x_odom,y_odom,theta_odom,v_r,v_l,count,state
    
    while count != 0:
        await node.wait_for_variables({"motor.left.speed"})
        await node.wait_for_variables({"motor.right.speed"})
        await node.wait_for_variables({"button.center"})
        await node.wait_for_variables({"button.forward"})
        await node.wait_for_variables({"prox.ground.delta"})

        
    
        motor_left = node["motor.left.speed"]
        motor_right = node["motor.right.speed"]
        button_center = node["button.center"]
        button_forward = node["button.forward"]
        prox_ground = node["prox.ground.delta"]

        x,y,theta = await odometrie(x,y,theta,motor_left,motor_right,0.1)
        #diff = prox_ground[1] - prox_ground[0]
        #print(prox_ground[0])
        x_odom.append(x)
        y_odom.append(y)
        theta_odom.append(theta)
        v_r.append(motor_right)
        v_l.append(motor_left)
        
        if button_forward == 1:
            state = True

        if state == True:
            count -= 1
            await node.set_variables(motors(50, -30))
            
        #count -= 1
        print('x odom: ', x, 'y odom: ', y, 'theta odom', theta)

        if button_center == 1 or count == 0:
            await node.set_variables(motors(0, 0))
            break

        # if button_forward == 1:
        #     await node.set_variables(motors(100, 100))


            # print(f"Button center state: {button_center}")
            # print(f"Button forward state: {button_forward}")

        #if prox_ground[0] <=900:
            #await node.set_variables(motors(0, 0))
            #break
        await client.sleep(0.1)
    #await node.set_variables(motors(0, 0))
await main()
```
